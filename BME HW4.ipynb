```python
# QUESTION 4A

i = 0
list_prob = []

while i in range(0, 20):
    i = i + 1
    prob = comb(20, i) * (0.17 ** i) * (0.83 ** (20 - i))
    list_prob.append(prob)

list_x = list(range(1, 21))

plt.bar(list_x, list_prob)
plt.xlabel('Number of Hurricanes')
plt.ylabel('Probability')
plt.title('PMF for Number of Hurricanes in Nashville')

a = np.array(list_x)
b = np.array(list_prob)

list_x_squared = np.array(list_x)**2

c = np.array(list_x_squared)

print(f'The expected value using PMF is: {sum(a*b)}')
print(f'The expected value using n*p is: {20 * 0.17}')
print('')
print(f'The variance using PMF is: {sum(c*b) - (sum(a*b)) ** 2}')
print(f'The variance using n*p*q is: {20 * 0.17 * 0.83}')
```

```python
# QUESTION 4B

n = 0
list_cum = []
cum = 0

while n in range(0, 20):
    n = n + 1
    cum = cum + comb(20, n) * (0.17 ** n) * (0.83 ** (20 - n))
    list_cum.append(cum)

plt.bar(list_x, list_cum)
plt.xlabel('Number of Hurricanes')
plt.ylabel('Cumulative Probability')
plt.title('CMF for Number of Hurricanes in Nashville')

print('See below for explanation.')
```

```python
# QUESTION 4B CONT.

j = 0
list_cdf = []

while j in range(0, 20):
    j = j + 1
    cdf = binom.cdf(j, 20, 0.17)
    list_cdf.append(cdf)
    
plt.bar(list_x, list_cdf)
plt.xlabel('Number of Hurricanes')
plt.ylabel('Cumulative Probability')
plt.title('CMF for Number of Hurricanes in Nashville')
print('The two bar graphs are identical.')
```

```python
# QUESTION 4C

p = 0
list_cdf_o = []

while p in range(0, 20):
    p = p + 1
    cdf_o = 1 - binom.cdf(p, 20, 0.17) + comb(20, p) * (0.17 ** p) * (0.83 ** (20 - p))
    list_cdf_o.append(cdf_o)
    
plt.bar(list_x, list_cdf_o)
plt.xlabel('Number of Hurricanes')
plt.ylabel('Cumulative Probability')
plt.title('CMF Upper for Number of Hurricanes in Nashville')
```
