```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
from math import sqrt
from scipy import stats
from matplotlib.ticker import FormatStrFormatter
```

```python
# QUESTION 1A

df = pd.read_excel('data (1).xlsx')

x = np.array(df['biceps circumference (cm)'])

mew = x.mean()
var = x.std() ** 2

ax1 = sns.countplot(x="biceps circumference (cm)", data=df)
plt.title('Frequency Distribution of Bicep Circumference')
plt.xticks(np.arange(0, 120, 10.0))
plt.show()

mean = (df['biceps circumference (cm)']).mean()
std = (df['biceps circumference (cm)']).std()
var = std ** 2

print(f'The mean of the bicep circumferences is: {mean}')
print(f'The variance of the bicep circumferences is: {var}')
```

```python
# QUESTION 1B

list_3 = []
list_3_s = []
i = 0

while i in range(0, 84):
    samp3 = np.random.permutation(x)[0:3]
    samp3_s = samp3.std() ** 2
    list_3_s.append(samp3_s)
    samp3 = samp3.mean()
    list_3.append(samp3)
    i = i + 1
    
xbar3 = sum(list_3) / len(list_3)
var3 = np.var(list_3)
mean_3s = sum(list_3_s) / len(list_3_s)
    
list_6 = []
list_6_s = []
j = 0

while j in range(0, 42):
    samp6 = np.random.permutation(x)[0:6]
    samp6_s = samp6.std() ** 2
    list_6_s.append(samp6_s)
    samp6 = samp6.mean()
    list_6.append(samp6)
    j = j + 1
    
xbar6 = sum(list_6) / len(list_6) 
var6 = np.var(list_6)
mean_6s = sum(list_6_s) / len(list_6_s)

list_12 = []
list_12_s = []
k = 0

while k in range(0, 21):
    samp12 = np.random.permutation(x)[0:12]
    samp12_s = samp12.std() ** 2
    list_12_s.append(samp12_s)
    samp12 = samp12.mean()
    list_12.append(samp12)
    k = k + 1
    
xbar12 = sum(list_12) / len(list_12)
var12 = np.var(list_12)
mean_12s = sum(list_12_s) / len(list_12_s)
    
ax2 = sns.countplot(x=list_3)
ax2.xaxis.set_major_formatter(FormatStrFormatter('%.2f'))
plt.title('Frequency Distribution of Bicep Circumference for Sample Size 3')
plt.xlabel('Bicep Circumference (cm)')
plt.xticks(np.arange(0, 74, 10.0))
plt.show()

print('The sample mean is: ', xbar3)
print('The sample variance is: ', var3)
```

```python
# QUESTION 1B Cont.

ax3 = sns.countplot(x=list_6)
ax3.xaxis.set_major_formatter(FormatStrFormatter('%.2f'))
plt.title('Frequency Distribution of Bicep Circumference for Sample Size 6')
plt.xlabel('Bicep Circumference (cm)')
plt.xticks(np.arange(0, 40, 5.0))
plt.show()

print('The sample mean is: ', xbar6)
print('The sample variance is: ', var6)
```

```python
# QUESTION 1B Cont.

ax4 = sns.countplot(x=list_12)
ax4.xaxis.set_major_formatter(FormatStrFormatter('%.2f'))
plt.title('Frequency Distribution of Bicep Circumference for Sample Size 12')
plt.xlabel('Bicep Circumference (cm)')
plt.xticks(np.arange(0, 21, 3.0))
plt.show()

print('The sample mean is: ', xbar12)
print('The sample variance is: ', var12)
```

```python
# QUESTION 1B Cont.

print('''The sample mean is roughly the same (32) regardless of the sample size. 
      The sample variance decreases as the sample size increases.''')
```

```python
# QUESTION 1C

x84 = np.linspace(1, 85, 84)
x42 = np.linspace(1, 43, 42)
x21 = np.linspace(1, 22, 21)

plt.scatter(x84, list_3, c='red', label='Sample Size 3')
plt.scatter(x42, list_6, c='blue', label='Sample Size 6')
plt.scatter(x21, list_12, c='green', label='Sample Size 12')

plt.axhline(y=mew, label='Population Mean')

plt.title('Scatter Plot of Varying Sample Means')
plt.ylabel('Sample Mean')
plt.xlabel('Sample Count')
plt.legend()
plt.show()

print('The population mean is the average of the points.')
```

```python
# QUESTION 1D

plt.figure(figsize=(10, 10))

plt.scatter(x84, list_3_s, c='red', label='Sample Size 3')
plt.scatter(x42, list_6_s, c='blue', label='Sample Size 6')
plt.scatter(x21, list_12_s, c='green', label='Sample Size 12')

plt.axhline(y=var, c='black', label='Population Variance')

plt.axhline(y=mean_3s, c='red', label='Mean Variance n=3')
plt.axhline(y=mean_6s, c='blue', label='Mean Variance n=6')
plt.axhline(y=mean_12s, c='green', label='Mean Variance n=12')

plt.axhline(y=var3, c='red', linestyle='dashed', label='Sample Variance n=3')
plt.axhline(y=var6, c='blue', linestyle='dashed', label='Sample Variance n=6')
plt.axhline(y=var12, c='green', linestyle='dashed', label='Sample Variance n=12')

plt.title('Scatter Plot of Varying Sample Means')
plt.ylabel('Sample Variance')
plt.xlabel('Sample Count')
plt.legend()
plt.show()

print('''As n increases from 3 to 12, the mean variance approaches the population variance, and 
the sample variance decreases.''')
```

```python
# QUESTION 1E

ratio1 = mean_3s / var3
ratio2 = mean_6s / var6
ratio3 = mean_12s / var12

print('The ratio for a sample size of 3 is: ', ratio1)
print('The ratio for a sample size of 6 is: ', ratio2)
print('The ratio for a sample size of 12 is: ', ratio3)
print('')
print('''The ratio for each sample is roughly equal to the sample size. 
The ratios are expected because variance of sample means divided by population variance is n.''')
```

```python
# QUESTION 2A

mew = 100
sigma = 15
var = sigma ** 2

rands = mew + sigma * np.random.randn(5000)

xbar = rands.mean()
s = rands.std()

print('The calculated sample mean is: ', xbar)
print('The calculated sample standard deviation is: ', s)
```

```python
# QUESTION 2B

x = np.linspace(0, 200, 5000)

y1 = stats.norm.pdf(x, mew, sigma)
y2 = stats.norm.pdf(x, xbar, s)

plt.plot(x, y1, label='Normal Distribution', color='red') # plots normal distribution N(100, 15^2)
plt.bar(x, y2, label='Probability Distribution', color='blue') # plot probability distribution for sample

plt.title('Probability Distribution and Normal Distribution of Tensile Strength')
plt.xlabel('Tensile Strength (MPa)')
plt.ylabel('Probability')
plt.legend()
plt.show()
```
