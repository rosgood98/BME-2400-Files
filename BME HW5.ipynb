```python
# QUESTION 1C

prob_125_given_faulty = stats.poisson.pmf(125, mu=1/8000 * 1000000)
prob_125_given_working = stats.poisson.pmf(125, mu=1/12000 * 1000000)

prob_125 = prob_125_given_faulty + prob_125_given_working

prob_faulty = 0.05 # given

prob_faulty_given_125 = prob_125_given_faulty / prob_125

print('''The probability of the sensor being faulty given there were 125 error bits is equal to the probability of 
      having 125 error bits giving it is a faulty sensor, times the probability of a faulty sensor, all divided by 
      the probability of having 125 error bits. This simplies to the probability of a faulty sensor and 125 error bits divided by the probability of 125 error bits.''')
print('')
print(f'The probability is: {prob_faulty_given_125}')
```
